```python
import pandas as pd
import seaborn as sns 
import matplotlib.pyplot as plt
```

```python
# Load the processed data 
df = pd.read_csv('../data/processed/sales_cleaned.csv')
```

```python
# Graph 1: Total Revenue by City 
plt.figure(figsize=(12,9))
sns.barplot(
    data=df, 
    x='city', 
    y='total_price', 
    estimator=sum, 
    errorbar=None 
)
plt.title('Total Sales Revenue by City')
plt.ylabel('Total Revenue ($)')

plt.savefig('../images/graph1_city_revenue.png', dpi=300, bbox_inches='tight')

plt.show()

```

```python
# Graph 2: Product Category Performance (Bar Chart)
plt.figure(figsize=(12,9))
sns.barplot(
    data=df,
    x='total_price',
    y='product_category',
    hue = 'product_category',
    estimator=sum, 
    errorbar=None,
    palette='viridis',
    legend=False
)
plt.title('Total Revenue by Product Category')
plt.xlabel('Total Revenue ($)')
plt.ylabel('Category')

plt.savefig('../images/graph2_customer_spending.png')

plt.show()
```

```python
# Graph 3: Customer Spending y Type & Gender 
plt.figure(figsize=(12,9))
sns.barplot(
    data=df, 
    x='customer_type', 
    y='total_price', 
    hue= 'gender',
    estimator=sum,
    errorbar=None
)
plt.title('Spending Behavior: Customer Type vs Gender')

plt.savefig('../images/graph3_product_performance.png')

plt.show()
```

```python
# Graph 4: Price vs Quantity Correlation
plt.figure(figsize=(12,9))
sns.scatterplot(
    data=df,
    x='unit_price',
    y='total_price',
    hue='customer_type',
    alpha=0.7
)
plt.title('Relationship between Unit Price and Total Spend')

plt.savefig('../images/graph4_price_correlation.png')

plt.show()
```
